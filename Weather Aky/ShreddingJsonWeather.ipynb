```python
#This notebook extracts data from the weather website in json form. There on it normalizes the data and displays in a table. 
from pandas import Series,DataFrame
import pandas as pd
import urllib, json
from pandas.io.json import json_normalize
```

```python
url = "http://api.openweathermap.org/data/2.5/forecast?q=London,uk&mode=json&appid=618ccea85118db5694a46720437ba07c"
response = urllib.urlopen(url)
data = json.loads(response.read())
l = [d for d in data['list']]
```

```python
t1 = json_normalize(l) #Normalizes all data except the weather columns. The weather columns has to be normalized seperately
#and is done from t3 onwards.
```

```python
t2 = t1[[('main.temp'),('main.temp_max'),('main.temp_min'),('main.humidity'),('rain.3h'),('wind.speed')]] 
#selecting the necessary columns from the normalized table above (t1)
```

```python
t3 = json_normalize(l, 'weather','dt_txt') #Normlizes weather column and displays with time
```

```python
t4 =t3[[('dt_txt'),('description')]] #Only displayes weather description and time, can be changed if needed
```

```python
result = pd.concat([t4, t2],axis=1, join='inner') #concatinates the two normalised tables with the required columns
```
